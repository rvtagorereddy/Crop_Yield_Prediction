##  Ideal value of pH for various crops 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
cropph=pd.read_csv('C:\\Users\\Admin\\Downloads\\cropph.csv')
cropph.head()
```

```python
cropph.columns
```

```python
cropph.describe()
```

```python
# Visualize the distribution of pH values
sns.countplot(data=cropph, x=' Ph')
plt.title('pH Distribution')
plt.show()

# Visualize the count of crops with ideal pH values
plt.figure(figsize=(10, 6))
sns.countplot(data=cropph, x='Crop')
plt.title('Crop Count with Ideal pH Values')
plt.xticks(rotation=90)
plt.show()
```

## Actual Crop Yield based on Various Factors

```python
cropvaries=pd.read_csv("C:\\Users\\Admin\\Downloads\\cropproductiononvariousfactors.csv")
cropvaries.head()
```

```python
cropvaries.columns
```

```python
# Get the summary statistics of the dataset
print(cropvaries.describe())
```

```python
# Check the data types of each column
print(cropvaries.dtypes)
```

```python
# Check for missing values
print(cropvaries.isnull().sum())
```

```python
# Visualize the distribution of 'Rainfall', 'Temperature', 'Ph', and 'Cropconversion'
sns.countplot(data=cropvaries, x='Rainfall')
plt.title('Rainfall Distribution')
plt.show()

sns.countplot(data=cropvaries, x='Temperature')
plt.title('Temperature Distribution')
plt.show()

sns.countplot(data=cropvaries, x='Ph')
plt.title('pH Distribution')
plt.show()

sns.countplot(data=cropvaries, x='Cropconversion')
plt.title('Crop Conversion Distribution')
plt.show()
```

```python
# Visualize the relationship between 'Rainfall' and 'Production' using a scatter plot
sns.scatterplot(data=cropvaries, x='Rainfall', y='Production')
plt.title('Rainfall vs. Production')
plt.show()
```

```python
# Visualize the average cost of production by crop
plt.figure(figsize=(10, 6))
sns.barplot(data=cropvaries, x='Crop', y='Production')
plt.title('Average Cost of Production by Crop')
plt.xticks(rotation=90)
plt.show()
```

## State Wise Cost of Production

```python
crop_yeild=pd.read_csv("C:\\Users\\Admin\\Downloads\\state_wise_crop_production.csv")
crop_yeild.head()
```

```python
crop_yeild.columns
```

```python
# Get the summary statistics of the dataset
print(crop_yeild.describe())

# Check the data types of each column
print(crop_yeild.dtypes)

# Check for missing values
print(crop_yeild.isnull().sum())
```

```python
# Visualize the distribution of'Yield (Quintal/ Hectare)'

sns.countplot(data=crop_yeild, x='Yield (Quintal/ Hectare) ')
plt.title('Crop Yield Distribution')
plt.show()
```

```python
# Visualize the relationship between 'Cost of Cultivation (`/Hectare) A2+FL' and 'Yield (Quintal/ Hectare)'
sns.scatterplot(data=crop_yeild, x='Cost of Cultivation (`/Hectare) A2+FL', y='Yield (Quintal/ Hectare) ')
plt.title('Cost of Cultivation (A2+FL) vs. Crop Yield')
plt.show()
```

```python
# Visualize the average yield by crop
plt.figure(figsize=(10, 6))
sns.barplot(data=crop_yeild, x='Crop', y='Yield (Quintal/ Hectare) ')
plt.title('Average Crop Yield by Crop')
plt.xticks(rotation=90)
plt.show()
```
